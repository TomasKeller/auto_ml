```python
from auto_ml import Predictor
#from auto_ml.utils import get_boston_dataset
from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(df[(df.app_type=='fl')&(df.app_granted_amount==5000)&(df.nibc_c_finance_credit_score < 20)][predictors].fillna(0), test_size=0.3)

column_descriptions = {
    'target': 'output',
    'adtraction_channel_url': 'categorical',
    'app_bank': 'categorical',
    'adtraction_campaign_type': 'categorical',
    'app_invoice_distribution_type': 'categorical',
    'adtraction_channel': 'categorical',
    'app_city_slugified': 'categorical',
    'app_city': 'categorical'
}

ml_predictor = Predictor(type_of_estimator='classification', column_descriptions=column_descriptions)

ml_predictor.train(df_train)

#ml_predictor.score(df_test, df_test.MEDV)
```

```python
#ml_predictor.train(df_train, ml_for_analytics=True,verbose=False)
ml_predictor.train(df_train, verbose=False)
```

```python
ml_predictor.train(df_train, feature_learning=True, fl_data=df_test)
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

y_pred = ml_predictor.predict_proba(df_test)[:, 1]
fpr, tpr, _ = roc_curve(df_test.target, y_pred)
auc = roc_auc_score(df_test.target, y_pred)
        
label = 'AUC = {:.2f}'.format(auc)

plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr, tpr, label=label)
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('ROC curve')
plt.legend(loc='best')
```
